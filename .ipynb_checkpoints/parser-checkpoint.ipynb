```python
import fitz  # PyMuPDF
import spacy
```

```python
# Buka file PDF
doc = fitz.open("resume.pdf")

# Ekstrak teks dari semua halaman
for page in doc:
    text += page.get_text()
    # print(text)

# Tutup dokumen
doc.close()
```

```python
print(f"parser {text}")
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
def preprocess_resume(text):
    text = text.replace("\n", "")         # Hilangkan newline
    text = text.replace("●", "")
    text = text.replace("|", "")
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
```

```python
preprocess_resume(text)
```

```python
jd_input = open("ai-engineer.txt", "r")
jd_text = jd_input.read()
jd_input.close()

print(jd_text)
```

```python
def preprocess_jd(jd_text):
    jd_text = jd_text.replace("\n", " ")  # ganti newline dengan spasi
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(jd_text)

    filtered_jdText = []
    for token in doc:
        if token.is_stop or token.is_punct or not token.is_alpha:
            continue
        filtered_jdText.append(token.lemma_.lower())  # lowercase untuk konsistensi

    return filtered_jdText

preprocess_jd(jd_text)
```

```python
preprocess_jd(jd_text)
```
